### add two features: holiday and weekday

```python
'''list out holidays between July to Oct'''
china_holidays_str = ['2016-09-15','2016-09-16','2016-09-17',
                      '2016-10-01','2016-10-02','2016-10-03',
                      '2016-10-04','2016-10-05','2016-10-06','2016-10-07']
china_holidays = []
for i in range(len(china_holidays_str)):
    china_holidays.append(datetime.strptime(china_holidays_str[i], "%Y-%m-%d").date())

'''special workding days after holidays'''
china_workingdays_str = ['2016-09-18', '2016-10-08', '2016-10-09']
china_workingdays = []
for i in range(len(china_workingdays_str)):
    china_workingdays.append(datetime.strptime(china_workingdays_str[i], "%Y-%m-%d").date())
```

```python
'''enconding feature: weekday; if it is working day, weekday=1 otherwise is 0'''
weekday = [0,1,2,3,4]
train_table['weekday'] = np.where((train_table['day_of_week'].isin(weekday))|(pd.to_datetime(train_table['date']).dt.date.isin(china_workingdays)),1,0)
train_table['weekday'].value_counts()
```

```python
'''enconding feature: holiday; if it is holiday, holiday=1 otherwise is 0'''
train_table['holiday'] = np.where(pd.to_datetime(train_table['date']).dt.date.isin(china_holidays),1,0)
train_table['holiday'].value_counts()
```

## To get road features

```python
links['link_id'] = links['link_id'].astype('int')
links
```

```python
'''add some columns to table4: road'''
road['link_list'] = road['link_seq'].str.split(',')
road['link_num'] = road['link_list'].map(lambda x: len(x))
road['road_length'] = 0
road['road_min_width'] = 0
road['total_in_top_num'] = 0
road['total_out_top_num'] = 0
```

```python
i=0
for index,row in road.iterrows():
    link_list=getattr(row,'link_list')
    road_length=0
    road_min_width=1000
    total_in_top_num=0
    total_out_top_num=0  
    
    for linkid in link_list:
        road_length+=int(links[links['link_id']==int(linkid)]['length'].values)
        total_in_top_num+=len(links[links['link_id']==int(linkid)]['in_top'].str.split(','))
        total_out_top_num+=len(links[links['link_id']==int(linkid)]['out_top'].str.split(','))
        temp=int(links[links['link_id']==int(linkid)]['width'].values)
        if temp<road_min_width:
            road_min_widt=temp
    road.loc[i,['road_length']]=road_length
    road.loc[i,['road_min_width']]=road_min_width
    road.loc[i,['total_in_top_num']]=total_in_top_num
    road.loc[i,['total_out_top_num']]=total_out_top_num
    
    i+=1
```

```python
road['route']=road['intersection_id']+'-'+road['tollgate_id'].astype('str')
```
