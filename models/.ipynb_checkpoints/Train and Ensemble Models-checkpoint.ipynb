# Feature ranking

```python
from sklearn.metrics import mean_squared_error as MSE
from sklearn.model_selection import learning_curve, validation_curve, GridSearchCV # Model evaluation
from xgboost import XGBRegressor,XGBClassifier,plot_importance # XGBoost
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold, cross_val_score, train_test_split # Model evaluation
from sklearn.preprocessing import StandardScaler,MinMaxScaler # Preprocessing
from sklearn.linear_model import Lasso, Ridge, RandomizedLasso, LinearRegression,ElasticNet, RANSACRegressor, SGDRegressor, HuberRegressor, BayesianRidge # Linear models
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier  # Ensemble methods
from sklearn.svm import SVR, SVC, LinearSVC  # Support Vector Regression
from sklearn.tree import DecisionTreeRegressor # Decision Tree Regression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline # Streaming pipelines
from sklearn.decomposition import KernelPCA, PCA # Dimensionality reduction
from sklearn.feature_selection import SelectFromModel # Dimensionality reduction
from sklearn.feature_selection import RFE, f_regression

colnames=training1.drop('route',axis=1).columns

###################################################################################
##########feature selection################
###################################################################################
# Define dictionary to store our rankings
ranks = {}
# Create our function which stores the feature rankings to the ranks dictionary
def ranking(ranks, names, order=1):
    minmax = MinMaxScaler()
    ranks = minmax.fit_transform(order*np.array([ranks]).T).T[0]
    ranks = map(lambda x: round(x,2), ranks)
    return dict(zip(names, ranks))

# Finally let's run our Selection Stability method with Randomized Lasso
rlasso = RandomizedLasso(alpha=0.04)
rlasso.fit(training1.drop('route',axis=1), training1_y['avg_travel_time_1'])
ranks["rlasso/Stability"] = ranking(np.abs(rlasso.scores_), colnames)
print('finished')


###################################################################################
##########feature selection################
###################################################################################
# Construct our Linear Regression model
lr = LinearRegression(normalize=True)
lr.fit(training1.drop('route',axis=1), training1_y['avg_travel_time_1'])
#stop the search when only the last feature is left
rfe = RFE(lr, n_features_to_select=1, verbose =3 )
rfe.fit(training1.drop('route',axis=1), training1_y['avg_travel_time_1'])
ranks["RFE"] = ranking(list(map(float, rfe.ranking_)), colnames, order=-1)

###################################################################################
##########feature selection################
###################################################################################
# Using Linear Regression
lr = LinearRegression(normalize=True)
lr.fit(training1.drop('route',axis=1), training1_y['avg_travel_time_1'])
ranks["LinReg"] = ranking(np.abs(lr.coef_), colnames)

# Using Ridge
ridge = Ridge(alpha = 7)
ridge.fit(training1.drop('route',axis=1), training1_y['avg_travel_time_1'])
ranks['Ridge'] = ranking(np.abs(ridge.coef_), colnames)

# Using Lasso
lasso = Lasso(alpha=.05)
lasso.fit(training1.drop('route',axis=1), training1_y['avg_travel_time_1'])
ranks["Lasso"] = ranking(np.abs(lasso.coef_), colnames)


```

```python
#XGBClassifier
xgb = XGBRegressor()
xgb.fit(training1.drop('route',axis=1), training1_y['avg_travel_time_1'])
ranks["XGB"] = ranking(xgb.feature_importances_, colnames)
```

```python
#RandomForestClassifier
rfc = RandomForestRegressor()
rfc.fit(training1.drop('route',axis=1), training1_y['avg_travel_time_1'].values)
ranks["RFClassifier"] = ranking(rfc.feature_importances_, colnames)
```

```python
#RandomForestClassifier
rfc = RandomForestRegressor()
rfc.fit(training1.drop('route',axis=1), training1_y['avg_travel_time_1'].values)
ranks["RFClassifier"] = ranking(rfc.feature_importances_, colnames)
```

```python
'''Create empty dictionary to store the mean value calculated from all the scores'''
r = {}
for name in colnames:
    
    r[name] = round(np.mean([ranks[method][name]
                             for method in ranks.keys()]), 2)

methods = sorted(ranks.keys())
ranks["Mean"] = r
methods.append("Mean")

print("\t%s" % "\t".join(methods))
for name in colnames:
    print("%s\t%s" % (name, "\t".join(map(str,
                                          [ranks[method][name] for method in methods]))))
```

```python
'''Put the mean scores into a Pandas dataframe'''
meanplot = pd.DataFrame(list(r.items()), columns= ['Feature','Mean Ranking'])

'''Sort the dataframe'''
meanplot = meanplot.sort_values('Mean Ranking', ascending=False)
```

```python
'''Let's plot the ranking of the features'''
sns.set(font_scale=2)
sns.factorplot(x="Mean Ranking", y="Feature", data = meanplot[0:10], kind="bar", 
               size=7, aspect=1.9, palette='coolwarm')
```

```python
class MultiOutput(torch.nn.Module):  # Two hidden layers neural network model.
    def __init__(self, num_hiddens1, num_hiddens2, num_out0,num_h1, num_h2, num_inputs=[9, 16, 16, 16, 16, 16, 16]):
        super(MultiOutput, self).__init__()
        self.hidden1 = torch.nn.Linear(num_inputs[0], num_hiddens1)
        self.hidden2 = torch.nn.Linear(num_hiddens1, num_hiddens2)
        self.out0 = torch.nn.Linear(num_hiddens2, num_out0)
        
        self.h11 = torch.nn.Linear(num_out0+num_inputs[1], num_h1)
        self.h12 = torch.nn.Linear(num_h1, num_h2)
        self.out1 = torch.nn.Linear(num_h2, 1)
        
        self.h21 = torch.nn.Linear(num_out0+num_inputs[2], num_h1)
        self.h22 = torch.nn.Linear(num_h1, num_h2)
        self.out2 = torch.nn.Linear(num_h2, 1)
        
        self.h31 = torch.nn.Linear(num_out0+num_inputs[3], num_h1)
        self.h32 = torch.nn.Linear(num_h1, num_h2)
        self.out3 = torch.nn.Linear(num_h2, 1)
        
        self.h41 = torch.nn.Linear(num_out0+num_inputs[4], num_h1)
        self.h42 = torch.nn.Linear(num_h1, num_h2)
        self.out4 = torch.nn.Linear(num_h2, 1)
        
        self.h51 = torch.nn.Linear(num_out0+num_inputs[5], num_h1)
        self.h52 = torch.nn.Linear(num_h1, num_h2)
        self.out5 = torch.nn.Linear(num_h2, 1)
        
        self.h61 = torch.nn.Linear(num_out0+num_inputs[6], num_h1)
        self.h62 = torch.nn.Linear(num_h1, num_h2)
        self.out6 = torch.nn.Linear(num_h2, 1)
        

    def forward(self, inputs):
        
        x = torch.index_select(inputs, 1, torch.tensor(range(0,num_inputs[0])))
        x = F.relu(self.hidden1(x))
        x = F.relu(self.hidden2(x))
        x = self.out0(x)

        
        y1 = torch.cat((x,torch.index_select(inputs, 1, torch.tensor(range(num_inputs[0], num_inputs[0]+num_inputs[1])))), dim=1)
        y1 = F.relu(self.h11(y1))
        y1 = F.relu(self.h12(y1))
        y1 = self.out1(y1)
        
        y2 = torch.cat((x,torch.index_select(inputs, 1, torch.tensor(range(num_inputs[0]+num_inputs[1], num_inputs[0]+num_inputs[1]+num_inputs[2])))), dim=1)
        y2 = F.relu(self.h11(y2))
        y2 = F.relu(self.h12(y2))
        y2 = self.out1(y2)
        
        y3 = torch.cat((x,torch.index_select(inputs, 1, torch.tensor(range(num_inputs[0]+num_inputs[1]+num_inputs[2], num_inputs[0]+num_inputs[1]+num_inputs[2]+num_inputs[3])))), dim=1)
        y3 = F.relu(self.h11(y3))
        y3 = F.relu(self.h12(y3))
        y3 = self.out1(y3)
        
        y4 = torch.cat((x,torch.index_select(inputs, 1, torch.tensor(range(num_inputs[0]+num_inputs[1]+num_inputs[2]+num_inputs[3], num_inputs[0]+num_inputs[1]+num_inputs[2]+num_inputs[3]+num_inputs[4])))), dim=1)
        y4 = F.relu(self.h11(y4))
        y4 = F.relu(self.h12(y4))
        y4 = self.out1(y4)
        
        y5 = torch.cat((x,torch.index_select(inputs, 1, torch.tensor(range(num_inputs[0]+num_inputs[1]+num_inputs[2]+num_inputs[3]+num_inputs[4], num_inputs[0]+num_inputs[1]+num_inputs[2]+num_inputs[3]+num_inputs[4]+num_inputs[5])))), dim=1)
        y5 = F.relu(self.h11(y5))
        y5 = F.relu(self.h12(y5))
        y5 = self.out1(y5)
        
        y6 = torch.cat((x,torch.index_select(inputs, 1, torch.tensor(range(num_inputs[0]+num_inputs[1]+num_inputs[2]+num_inputs[3]+num_inputs[4]+num_inputs[5], num_inputs[0]+num_inputs[1]+num_inputs[2]+num_inputs[3]+num_inputs[4]+num_inputs[5]+num_inputs[6])))), dim=1)
        y6 = F.relu(self.h11(y6))
        y6 = F.relu(self.h12(y6))
        y6 = self.out1(y6)
        

        b, h = inputs.shape 



        return torch.cat((y1,y2,y3,y4,y5,y6),dim=1).view(b,-1)
```

### define mape 

```python
def MAPE(result):
    mape=0
    routes=list(set(result['route'].values))
    R=len(routes)
    
    for route in routes:
        tws=result[result['route']==route][['real','pred']].values
        T=0
        ma=0
        for tw in tws:
            ma+=abs((tw[0]-tw[1])/tw[0])
            T+=1
            #print(tw)
        mape+=ma/T
        R+=1
    mape/=R
        
    return mape
```

```python
dt = df['time_window_2h_start']
df['first_20min_period_start'] = dt.map(lambda x: x+h2_delta)
df['second_20min_period_start'] = dt.map(lambda x: x+h2_delta+min_delta)
df['third_20min_period_start'] = dt.map(lambda x: x+h2_delta+2*min_delta)
df['forth_20min_period_start'] = dt.map(lambda x: x+h2_delta+3*min_delta)
df['fifth_20min_period_start'] = dt.map(lambda x: x+h2_delta+4*min_delta)
df['sixth_20min_period_start'] = dt.map(lambda x: x+h2_delta+5*min_delta)
```

```python
df
```

```python
df = df.iloc[:,1:]
```

```python
first_20min_pred = df[['route','first_20min_period_start','p1']]
second_20min_pred = df[['route','second_20min_period_start','p2']]
third_20min_pred = df[['route','third_20min_period_start','p3']]
forth_20min_pred = df[['route','forth_20min_period_start','p4']]
fifth_20min_pred = df[['route','fifth_20min_period_start','p5']]
sixth_20min_pred = df[['route','sixth_20min_period_start','p6']]
```

```python
first_20min_pred
```

```python
first_20min_pred = first_20min_pred.rename(columns={'first_20min_period_start':'time_window_start','p1':'avg_travel_time'})
second_20min_pred = second_20min_pred.rename(columns={'second_20min_period_start':'time_window_start','p2':'avg_travel_time'})
third_20min_pred = third_20min_pred.rename(columns={'third_20min_period_start':'time_window_start','p3':'avg_travel_time'})
forth_20min_pred = forth_20min_pred.rename(columns={'forth_20min_period_start':'time_window_start','p4':'avg_travel_time'})
fifth_20min_pred = fifth_20min_pred.rename(columns={'fifth_20min_period_start':'time_window_start','p5':'avg_travel_time'})
sixth_20min_pred = sixth_20min_pred.rename(columns={'sixth_20min_period_start':'time_window_start','p6':'avg_travel_time'})
```

```python
first_20min_pred
```

```python
submit = pd.concat([first_20min_pred,second_20min_pred,third_20min_pred,forth_20min_pred,fifth_20min_pred,sixth_20min_pred])
```

```python
submit.sort_values(by=['route','time_window_start'], inplace=True)
submit = submit.reset_index(drop=True)
```

```python
submit
```

```python
routes = submit['route']
submit['intersection_id'] = routes.map(lambda route: route[0])
submit['tollgate_id'] = routes.map(lambda route: route[2])
```

```python
time_window_start = submit['time_window_start']
submit['time_window_end'] = time_window_start.map(lambda x: x+min_delta)
```

```python
submit['time_window_start'] = submit['time_window_start'].astype('str')
submit['time_window_end'] = submit['time_window_end'].astype('str')
submit['time_window'] = '['+submit['time_window_start']+','+submit['time_window_end']+')'
```

```python
submission = submit[['intersection_id','tollgate_id','time_window','avg_travel_time']]
```
